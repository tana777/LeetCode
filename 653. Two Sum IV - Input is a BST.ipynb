##### 653. Two Sum IV - Input is a BST

Given a Binary Search Tree and a target number, return true if there exist two elements in the BST such that their sum is equal to the given target.

Example 1:
Input: 
                 
                 5
                / \
               3   6
              / \   \
             2   4   7

Target = 9

Output: True
Example 2:
Input: 
   
         5
        / \
       3   6
      / \   \
     2   4   7

Target = 28

Output: False

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


## BFS iteration + dictionary
class Solution(object):
    def findTarget(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: bool
        """
        if root is None:
            return False

        que = collections.deque()
        output = []
        num = 0
        que.append(root)
        while que:
            num = len(que)
            for i in range(num):
                node = que.popleft()
                if node:
                    output.append(node.val)
                    for j in (node.left,node.right):
                        que.append(j)
        dic = dict.fromkeys(output)
        for i in dic:
            if k-i in dic and k-i!=i:
                return True
        return False
               
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

## inorder traversal DFS + stack
## after get the ordered list then check if k-val in the list
class Solution(object):
    def findTarget(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: bool
        """
        if root is None:
            return False
        stack = []
        output = []
        while root or stack:
            if root:
                stack.append(root)
                root = root.left
            else:
                node = stack.pop()
                output.append(node.val)
                root = node.right
        for i in output:
            if k-i in set(output) and k-i!=i:
                return True
        return False
                
```
